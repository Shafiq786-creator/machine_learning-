# Principal Component Analysis (PCA)

 - ### What is PCA? 
 A technique used to reduce dimension while preserving essentail data characteristics.

 Helps in visualization, computational efficiency, and noise reduction

 - ### Why do we need PCA?
 High-dimensional data is complex and computationally expensive 

 PCA find the most important features and removes redundant information

## Curse of Dimensionality

 - The curse of dimensionality in machine learning refers to various problems that arise when working with data in high-dimensional spaces(ie... With many features or variable )

 - The curse of dimensionality is the set of problems that arise as data dimensionlaity increases, including data sparsity, ineffectiv distace meteric, high computation , and overfitting, all of which can hurt machine learning performance

 - As the number of dimensions increases, the data becomes sparser and many algoriths become less effective or computationally expensive. 

## Understanding Data Variations

 1. Data variation is the spread or dispersion in data

 2. Why is variation important?

   - high Variation --important information

   - low variation -- can be ignored

 3. PCA focuses on maximizing variance to capture key patterns. 

## What happens in Higher Dimensions?

 1. ### 1D Data
  - single numerical axis(e.g price)

 2. ### 2D Data
  - Data points are scatterd across an x-y coordinate system(i.e price vs area )

 3. ### 3D Data
  - Data exists in three dimensions, but patterns may still lie on a 2D plane. 

## Finding principal Components

 1. The first principal component(PC1):

  captures the maximum variation

 2. The Second principal component(PC2):

  is perpendicular to PC1 and captures the second-most variation

 3. The process continues:
 
  For more dimensions  

## How are principal Components Computed? 

uses linear algebra concepts to extract meaningful components from complex data.

- **Eigenvectors & Eigenvalues**: 
Mathematical foundations of PCA 

- **Singular Value Decomposition(SVD)**:
Computational technique

- **Direction Finding**: 
Find the direction that capture the most variation


SVD factorizes a matrix A into three matrices:

                A = UEVt

Where:
 - A is the original m x n matrix (e.g, data matrix)
 - U is an m x m matrix of left singular vectors (orthonormal columns)
 - E is an m x n diagonal matrix of singular values(non-negative real numbers).
 - Vt is the transpose of an n x n matrix of right singular vectors (orthonormal rows).                

## Singular Value Decomposition (SVD)

Mathematical method to extract principal components. 

 1. **Compute covariace matrix** : Calculate how variables relate to each other 

 2. **Find eignvectors and eigenvalues** : identify key directions and their importance

 3. **Rank Components based on importance** : order by variance explained

## Sorting Principal Components

 1. **Ranking by sigma values**
  
   Components are sorted based on their singular values (sigma values)

 2. **Significance Hierarchy**

   Larger sigma value -- more significant component

 3. **Dimension Reduction**

   Lower sigma values can be ignored.     

## Steps to Apply PCA

- step 1 : Standardize the dataset.
- step 2 : Compute the covariance matrix
- step 3 : Perform Eigen decomposition
- step 4 : Select the top components
- step 5 : Transform the dataset

## Benefits of PCA 

 - Removes redundant data.
 - Speed up machine learning models.
 - imporves visualization by reducing dimensions.
 - Reduces noise by eliminating unimportant variations.


## Limitations of PCA

 - **Linear Relationships**: PCA assumes linear relationship, non_linear data may not be well represented. 

 - **Interpretability Issues** : principal components are abstract

 - **Scale Sensitivity** : Data should be standarized

## PCA vs. Traditional Feature Selection

### Feature Selection

 - Removes features based on statistical tests
 - May discard important correlated features
 - Works well for categorical data


### PCA

 - Transform data into a new coordinate system
 - Retains maximum variance
 - Best for numerical data


## Applications of PCA:

- PCA is applied in various fields, such as:

  - **Image Processing**: To reduce the number of pixels in images while retaining key features.
  - **Finance**: To identify the principal factors affecting market trends.
  - **Genetics**: For gene expression data analysis.
  - **Speech Recognition**: To capture critical acoustic features of speech.

## Example of PCA Implementation

### Import Required Libraries
We first import the necessary libraries to perform PCA and visualization:

 - **numpy**: Used for numerical operations.
 - **matplotlib.pyplot**: Used for creating visualizations like scatter plots.
 - **load_iris and load_wine**: Functions from sklearn.datasets to load the Iris and Wine datasets.
 - **StandardScaler**: From sklearn.preprocessing, used to standardize the data.
 - **PCA**: From sklearn.decomposition, used to apply Principal Component Analysis.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# step 1: Load the iris dataset
iris = load_iris()
X = iris.data  # features
y = iris.target # lables

target_names = iris.target_names

# step 2: standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# step 3: Apply pca to reduce to 2 components
pca = PCA(n_components=2)
x_pca = pca.fit_transform(X_scaled)

# step 4: Visualize the 2D PCA result
plt.figure(figsize=(8, 6))
colors = ['navy', 'turquoise', 'darkorange']
for color , i, target_name in zip(colors, [0, 1, 2],target_names ):
    plt.scatter(x_pca[y ==i, 0], x_pca[y ==i, 1], alpha=0.7, color=color, label=target_name)

plt.legend()
plt.title("PCA or IRIS dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("principal Component 2")
plt.grid(True)
plt.tight_layout()
plt.show()    

# Step 5 : Explained Variance
print("Explained Variance ratio: ", pca.explained_variance_ratio_)
```

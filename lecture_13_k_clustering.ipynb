## What is Clustering?
**Clustering** is the process of dividing data points into distinct groups (clusters) based on some **distance** or **similarity** measures. The task of grouping datapoints based on their similarity with each other is also called clustering or cluster analysis. The core idea is to group similar data points together while separating dissimilar points.

## How K-means Clustering Works?

 - **Initialize** : Select initial points k called centroids.
 - **Assign** : Assign each data point to the nearest centroid based on a distance measure.
 - **Update** : Recalculate centroids based on the assigned points.
 - **Repeat** : Repeat the assignment and centroid update process until convergence (no change in clusters).

## Distance and Similarity Measures Used in Clustering
Choosing an appropriate distance or similarity metric is crucial for effective clustering. The common metrics include:

 1. **Euclidean Distance**

Used when data points follow a bell-shaped (Gaussian) distribution.
Measures the straight-line distance between two points in space.
  - **Explaining**
  The Euclidean distance is the most widely used distance measure in clustering. It calculates the straight-line distance two points in n-dimensional space.
  The formula of Euclidean distance between two points A(x1, y1) and B(x2, y2)is 
  
           d(A,B) = squrroot[(x2-x1)2 + (y2-y1)2]

   The Euclidean distance always measures the shortest possible path between two points, represented by a straight line.

 2. **Cosine Similarity**

Used primarily for textual data and Natural Language Processing (NLP).
Measures the cosine of the angle between two vectors, reflecting their orientation similarity regardless of magnitude.
Measures angle between vectors, ideal for high-dimentional data
  - **Explaning**
  Measure the cosine of the angle between two vectors, focusing on orientation rather than magnitude
    - **formula**
      
      cos(0) = (A-B)/(||A||-||B||)
    - **Defination**

    Calculates alignment between vectors by finding the cosine of their angle

    - **Key Property**

    Higher Values (closer to 1) indicate greater similarity between vectors

    - **Appliction**

    Ideal for text analysis, document, comparison, and recomendation systems   

    - **Advantage**

    Ignore magnitude differences, focusing on directional similarity

 3. **Manhattan Distance**

Applied when data has a uniform distribution.
Measures distance along axes at right angles (like navigating city blocks).Sum of absolute differences between points useful for grid like structures
 - Measures the total absolute difference between the cartesian coordinates of two points
  - **Analogy** : like moving across a city grid . where only horizontal and verticle moves are allowed.
  - **formula** : Sum of the absolute difference in each dimension

      distance = summation|xi - yi| for all dimensions i

 4. **Minkowski Distance**

A generalization of Euclidean and Manhattan distances.
q=1q=1 is Manhattan Distance
q=2q=2 is Euclidean Distance

```python
# import Required libraries
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the Iris datasets
iris = load_iris()
X = iris.data  # features
y_true = iris.target # actual labels (used only for comparison, not clustering)


# Create a DataFrame for better visualization
df = pd.DataFrame(X, columns=iris.feature_names)
df['Actual_Label'] = y_true

# Apply Kmeans Clustering
k = 3  # we know iris has 3 natural clusters
kmeans = KMeans(n_clusters=k, random_state=42)
df['Cluster_Label'] = kmeans.fit_predict(X)

#optional: visualize clustering in 2D using first two features
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X[:,0], y=X[:,1], hue=df['Cluster_Label'], palette ='viridis', s=100)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("K-Means Clustering(2 Features)")
plt.legend()
plt.grid(True)
plt.show()
```

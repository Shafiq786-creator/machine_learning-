# Class Imbalance

## What is Class Imbalance?
A situation where one class is represented much more frequently than others in a dataset.

## Why it Matters? 
Models can acheieve misleadingly high accuracy by simply predicting the majority calss

## Impact on Model Performance
Can lead to poor generalization and unreliable predictions on real world data.


## Accuracy Example with Class Imbalance

1 --- **Training Set**
          
           98% samples of class A and 2% samples of class B. MOdel achieve 98% training accuracy by simply predicting every sample as class

2 --- **Test Set**

           60% samples of class A and 40% samples of class B
           test accuracy drops down to 60%

3 --- **Problem**

           Classification accuracy gives us the false sense of achieveing high accuracy 
           the real problem arises when the cost of misclassification of the minor class samples are very high           

## Confusion Matrix

 - **Definition**
   Confusion Matrix as the name suggests gives us a matrix as output and describe the complete performance of the model

 - **Purpose**
   provides a detailed breakdown of predictions versus actual values

 - **Adavantage** 
   Allows calculation of various performance metrics beyond simple accuracy    

## Components of Confusion Matrix

 - **True Positive**

   The cases in which we predicted YES and the actual output was also YES

 - **fasle Positive**

   The cases in which we predicted YES and the actual output was also NO

 - **True Neagative**

   The cases in which we predicted NO and the actual output was also NO

 - **False Negative**

   The cases in which we predicted NO and the actual output was also YES

## ROC Curve (Receiver Operating Characteristic)

 - **Definition**

         A graph showing the trade-off betwween true positvie rate (recall) and flase positive rate at different thresholds.

 - **Purpose**

         Helps visualize model performace across different threshold values

 - **True Positive Rate**(sensitivity)

         Proportion of actual positives correctly identified by the model.

 - **False Posiive Rate**(specificity)

         proportion of actual negatives incorrectly identified as positives by the model.                        

## AUC (Area Under the Curve)

 - **perfect model: AUC = 1.0**
           Perfect separation between classes


 - **Good model: AUC>0.8**
           Strong discrimination ability


 - **Random Model : AUC = 0.5**
           No better than random guessing 


Area Under the ROC curve: Measures the model's ability to distinguish between the classed

summarized the model's overall ability to discriminate between positive and negative classes.

## Mean Squared Error (MSE)
 - **Definition**

   Average of squared differences between predicted and actual values.

 - **Characteristics**

   Penalized large errors due to squaring, making it sensitive to outliers. Always non-negative with values closer to zero indicating better performance.


            MSE= 1/n ∑(y-y^)2



## Mean Absolute Error (MAE)
 - **Definition**

   Average of absolute differences between predicted and actual values.Measures error magnitude regardless of directions

 - **Comparison**

   More robust to outliers than MSE as it uses absolute rather than squared difference. Treats errors linearly instead of quadratically, making it less sensitive to occasional large errors and better suited for noisy data.


            MAE= 1/n ∑(y-y^)



## Roor Mean Square Error (RMSE)
 - **Definition**

   Root Mean Square Error: The square root of MSE, providing error measurements in the same units as the target variable.

 - **Comparison**
     
      More sensitive to outliers than MAE and commonly used in statistical modeling


            RMAE= sqrt1/n ∑(y-y^)



## Mean Absolute Percentage Error (MAPE)
 - **Definition**

   Mean Absolute Percentage Error: Express accuracy as a percentage of error, making it scale-independdant and easily interpretable. 

 - **Comparison**
     
      Calculated as the average of absolute percentage difference between predicted and actual values.


            MAPE= (100%/n) ∑(y-y^)/y_i



## **Import Labraries**: 

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report, roc_curve, auc
)

import matplotlib.pyplot as plt
import seaborn as sns

```

### **Load the Iris Dataset**

```python
iris = load_iris()
X = iris.data
y = iris.target
```

### **Convert to binary classification**

```python
y_binary = (y==0).astype(int)
```

### **Train/test split**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### **Standardize the feature**

```python
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

### **Train Model**

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

### **Make Predictions**

```python

y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]
```

### **Evaluation** :

```python

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")

# Classification report
print("\nClassification Report: ")
print(classification_report(y_test, y_pred))

# Confusion Matrix
print("\nConfusion Matrix: ")
print(confusion_matrix(y_test, y_pred))

# roc auc
```
